```python
from MolAdsPy import Molecule
from MolAdsPy import Slab
from MolAdsPy import Adsorption
from numpy import array
from MolAdsPy import Atom



graphene=Slab("graphene", "graphene.xyz")
adsorbed=Adsorption(graphene)
adsorbed.resize(4,4)

h2_mol=Molecule("h2")

print(h2_mol)

for atom in h2_mol:
    print("Atom ",atom.ID,atom.coords)

print("\n2)")

h1=Atom("H",x=[0.5,0.0,0.0])
h2=h1.copy()
h2.coords[0]=0.746+0.5

h2_mol.add_atom(h1)
h2_mol.add_atom(h2)
h2_mol.write_xyz("h2.xyz")

print(len(graphene))

i=0
#h2_mol.rotate(0,0,-30)

for n in range(int(len(graphene)-1)):
    i+=1
    print(n)
    
    topN_pos=array([graphene[n].coords[0],graphene[n].coords[1]])

    bridgeN_pos=array([(graphene[n].coords[0]+(graphene[n+1].coords[0]-
                                               graphene[n].coords[0])/2.0),(graphene[n].coords[1]+(graphene[n+1].coords[1]-
                                               graphene[n].coords[1])/2.0)])
    
    graphene.add_adsorption_site("topN",topN_pos)    
    graphene.add_adsorption_site("bridgeN",bridgeN_pos)
    #h2_mol.rotate(0,0,30)
    adsorbed.add_component(h2_mol,"topN",n=2,m=2,anchor="com",vertsep=2.0,side="top")
    name_file='adsorbed'+str(i)+'.h2.xyz'
    print(name_file)
    adsorbed.write_xyz(name_file)
    i+=1
    adsorbed.remove_component(h2_mol)
    adsorbed.add_component(h2_mol,"bridgeN",n=2,m=2,anchor="com",vertsep=2.0,side="top")
    #h2_mol.rotate(0,0,30)
    name_file='adsorbed'+str(i)+'.h2.xyz'
    print(name_file)
    adsorbed.write_xyz(name_file)
    adsorbed.remove_component(h2_mol)
    
    



```
